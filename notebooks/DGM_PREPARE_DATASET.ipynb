```python
from datasets import load_dataset

dataset = load_dataset("ted_multi")
```

```python
train_set = dataset['test']
```

```python
def check_elements(list1, list2):
    for element in list1:
        if element not in list2:
            print(element)
            return False
    return True
```

```python
common_languages = ['hi','tr','pt','hu','id','en','es']
```

```python
import pandas as pd

data = {}

# Create an empty list for each column name
for column_name in common_languages:
    data[column_name] = []

# Create the dataframe from the data dictionary
df = pd.DataFrame(data)
```

```python
from googletrans import Translator

def translate_english_to_hindi(text):
    translator = Translator(service_urls=['translate.google.com'])
    translation = translator.translate(text, src='en', dest='hi')
    return translation.text

# Example usage
english_text = "Hello, how are you?"
hindi_text = translate_english_to_hindi(english_text)
print(hindi_text)
```

```python
counter = 0

while(counter != 100):
      
      lang = train_set['translations'][counter]['language']      
      transaltations = train_set['translations'][counter]['translation']
      temp = {}

      for l in common_languages:
        if(l in lang):
          index = lang.index(l)
          trans_temp = transaltations[index]
          temp[l] = trans_temp

      df = df.append(temp, ignore_index=True)
      counter = counter + 1
```

```python
translator = Translator(service_urls=['translate.google.com'])
```

```python
for index, row in df.iterrows():

  hi = row['hi']
  tr = row['tr']
  pt = row['pt']
  hu = row['hu']
  id = row['id']
  en = row['en']
  es = row['es']

  if pd.isnull(en):
    print("English is Null")

  else:

    text = en

    if(pd.isnull(hi)):
        translation = translator.translate(text, src='en', dest='hi')
        df.at[index, 'hi'] = translation.text
    if(pd.isnull(tr)):
        translation = translator.translate(text, src='en', dest='tr')
        df.at[index, 'tr'] = translation.text
    if(pd.isnull(pt)):
        print(text)
        translation = translator.translate(text, src='en', dest='pt')
        df.at[index, 'pt'] = translation.text
    if(pd.isnull(hu)):
        translation = translator.translate(text, src='en', dest='hu')
        df.at[index, 'hu'] = translation.text
    if(pd.isnull(id)):
        translation = translator.translate(text, src='en', dest='id')
        df.at[index, 'id'] = translation.text
    if(pd.isnull(es)):
        translation = translator.translate(text, src='en', dest='es')
        df.at[index, 'es'] = translation.text
```
